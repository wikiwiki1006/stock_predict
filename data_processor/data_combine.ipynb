```python
import pandas as pd
detail_data_path = '/Users/shin-yoonhwan/vscode/stock_data/data/detail_data_533.csv'
all_money_info_path = '/Users/shin-yoonhwan/vscode/stock_data/data/all_money_info.csv'

detail_data_df = pd.read_csv(detail_data_path)
all_money_info_df = pd.read_csv(all_money_info_path)
print(detail_data_df.shape)
print(all_money_info_df.shape)
```

```python
combined_data_df = pd.merge(detail_data_df, all_money_info_df, on = ['Ticker', 'Quarter'], how = 'left')
combined_data_df.shape
```

```python
combined_data_df.head()
```

```python
combined_data_df.to_csv('combined_data.csv')
```

```python
market_data_path = '/Users/shin-yoonhwan/vscode/stock_data/data/repeated_market_data.csv'
market_data_df = pd.read_csv(market_data_path)
market_data_df.shape
market_data_df = market_data_df.drop(columns= 'Unnamed: 0')
```

```python
all_combined_data = pd.concat([combined_data_df, market_data_df], axis = 1, ignore_index=False)
all_combined_data.head()
```

```python
all_combined_data.to_csv('all_combined_data.csv')
```
