```python
all_df_copy['index']= pd.to_datetime(all_df_copy['index'])

# 사용자 정의 함수: 월을 기반으로 분기 계산
def custom_quarter(date):
    year = str(date.year)[-2:] 
    if date.month in [10, 11, 12]:
        quarter = 3
    elif date.month in [7, 8, 9]:
        quarter = 2
    elif date.month in [4, 5, 6]:
        quarter = 1
    else:  # 1, 2, 3
        quarter = 4
        year = int(year) - 1
    return f"{year}.Q{quarter}"

# 인덱스를 사용자 정의 분기 형식으로 변환
all_df_copy['index'] = all_df_copy['index'].map(custom_quarter)
```
